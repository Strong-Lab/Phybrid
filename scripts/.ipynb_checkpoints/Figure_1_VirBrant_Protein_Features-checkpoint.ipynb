```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kruskal
```

## Protein Feature Figures

### 10K Contigs

```python
bac = pd.read_csv('../data/VirBrant_Training_Data/Protein_Features/Bac10k_AA_Features.csv')
viral = pd.read_csv('../data/VirBrant_Training_Data/Protein_Features/Phage10k_AA_Features.csv')
viral["Type"] = "Viral"
bac["Type"] = "Bacterial"
df_combined = pd.concat([viral, bac])

fig = plt.figure(figsize = (20, 12))

ax1 = fig.add_subplot(221) #top left
ax2 = fig.add_subplot(222) #top right
ax3 = fig.add_subplot(223) #bottom left
ax4 = fig.add_subplot(224) #bottom right

ax1.annotate("A", xy=(-0.1, 1.05), xycoords="axes fraction", fontsize=20, weight="bold")
ax2.annotate("B", xy=(-0.1, 1.05), xycoords="axes fraction", fontsize=20, weight="bold")
ax3.annotate("C", xy=(-0.1, 1.05), xycoords="axes fraction", fontsize=20, weight="bold")
ax4.annotate("D", xy=(-0.1, 1.05), xycoords="axes fraction", fontsize=20, weight="bold")



ax = sns.boxplot(x=df_combined["Type"], y=df_combined['Gene_Density'], palette="Set3", showfliers=False, ax=ax1)
ax.set_title("")
ax.set_xlabel("")
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=14)
ax.set_ylabel("Gene Density (Genes per 1K Bases)", fontsize=16)

ax = sns.boxplot(x=df_combined["Type"], y=df_combined['Median Operon Length'], palette="Set3", showfliers=False, ax=ax2)
ax.set_title("")
ax.set_xlabel("")
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=14)
ax.set_ylabel("Median Operon Length", fontsize=16)


ax = sns.boxplot(x=df_combined["Type"], y=df_combined['Percent_Overlap'], palette="Set3", showfliers=False, ax=ax3)
ax.set_title("")
ax.set_xlabel("")
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=14)

ax.set_ylabel("Percentage of Genes Overlapping", fontsize=16)


ax = sns.boxplot(x=df_combined["Type"], y=df_combined['AA_Length'], palette="Set3", showfliers=False, ax=ax4)
ax.set_title("")
ax.set_xlabel("")
ax.xaxis.set_tick_params(labelsize=20)
ax.yaxis.set_tick_params(labelsize=14)


ax.set_ylabel("Median Amino Acid Length", fontsize=16)

fig.savefig("../images/Protein_Feature_Boxplots/Proteins_10k.jpg")
```

```python
print("Gene Density")
print(kruskal(viral['Gene_Density'], bac['Gene_Density']))
print("--------")
print("Median Operon Length")
print(kruskal(viral['Median Operon Length'], bac['Median Operon Length']))
print("--------")
print("Percentage of Genes Overlapping")
print(kruskal(viral['Percent_Overlap'], bac['Percent_Overlap']))
print("--------")
print("Median Amino Acid Length")
print(kruskal(viral['AA_Length'], bac['AA_Length']))
```
